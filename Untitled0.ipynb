```python
"""
Water Quality Checker (Basic)

Features:
- User inputs: pH, Hardness (mg/L), TDS (mg/L)
- Basic quality check based on standard ranges
- Shows if water is Good / Acceptable / Poor
- Can be CLI (Python) or simple HTML form

Perfect small project for lab analytics impression in internships.
"""

# Python CLI version

def water_quality_check():
    print("--- Water Quality Checker ---")
    try:
        ph = float(input("Enter pH value: "))
        hardness = float(input("Enter Hardness (mg/L): "))
        tds = float(input("Enter TDS (mg/L): "))
    except ValueError:
        print("Invalid input. Please enter numeric values.")
        return

    # Simple quality rules (example)
    quality = "Good"
    if not (6.5 <= ph <= 8.5):
        quality = "Poor"
    elif not (0 <= hardness <= 200):
        quality = "Poor"
    elif not (0 <= tds <= 500):
        quality = "Poor"

    print(f"Water Quality Status: {quality}")


if __name__ == "__main__":
    water_quality_check()
```
